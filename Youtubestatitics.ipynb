```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = '/content/youtubers_df.csv'
youtubers_df = pd.read_csv(file_path)
```

```python
youtubers_df['Categories'].fillna('Unknown', inplace=True)
```

```python
youtubers_df.isnull().sum()
```

```python
youtubers_df['Rank'] = pd.to_numeric(youtubers_df['Rank'], errors='coerce')
```

```python
sns.set_style("whitegrid")
```

```python
top_5_subscribers = youtubers_df.sort_values(by='Suscribers', ascending=False).head(5)
plt.figure(figsize=(10, 6))
sns.barplot(x='Username', y='Suscribers', data=top_5_subscribers)
plt.title('Top 5 YouTube Channels by Subscribers')
plt.xlabel('YouTube Channel')
plt.ylabel('Subscribers')
plt.show()
```

```python
top_countries = youtubers_df['Country'].value_counts().head(10)  # Top 10 countries
plt.figure(figsize=(10, 8))
plt.pie(top_countries, labels=top_countries.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Channels in Top 10 Countries')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(youtubers_df['Likes'], bins=30, kde=True)
plt.title('Distribution of Likes Across Channels')
plt.xlabel('Likes')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Suscribers', y='Likes', data=youtubers_df)
plt.title('Relationship Between Subscribers and Likes')
plt.xlabel('Subscribers')
plt.ylabel('Likes')
plt.show()
```

```python
import matplotlib.pyplot as plt

top_countries = youtubers_df['Country'].value_counts().head(10)

plt.figure(figsize=(10, 6))
plt.barh(top_countries.index, top_countries, color='skyblue')
plt.xlabel('Number of Channels')
plt.ylabel('Country')
plt.title('Distribution of Channels in Top 10 Countries')
plt.show()
```
